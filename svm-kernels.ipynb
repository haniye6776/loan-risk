```python
import pandas as pd
from google.colab import drive
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.under_sampling import RandomUnderSampler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score, log_loss
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
drive.mount('/content/gdrive')
data = pd.read_csv('gdrive/My Drive/Datasetjadid.csv')
```

```python
data
```

```python
if data.duplicated().any():
    print(data.duplicated().sum(),"has duplicate data.")
else:
    print("no duplicate.")
```

```python
X = data.drop(['Risk_Flag','CITY','STATE'],axis=1)
y = data['Risk_Flag']
```

```python
X = pd.get_dummies(X,drop_first=True)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.isnull().sum()
```

```python
X_test.isnull().sum()
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X_train_scaled, y_train)
```

```python
dt_gini = DecisionTreeClassifier(criterion='gini', random_state=42)
dt_gini.fit(X_resampled, y_resampled)
y_pred_gini = dt_gini.predict(X_test_scaled)
y_pred_proba_gini = dt_gini.predict_proba(X_test_scaled)
print("\n'gini':")
cm_gini = confusion_matrix(y_test, y_pred_gini)
print(f"Confusion Matrix:\n{cm_gini}")
precision_gini = precision_score(y_test, y_pred_gini)
recall_gini = recall_score(y_test, y_pred_gini)
f1_gini = f1_score(y_test, y_pred_gini)
print(f"Precision: {precision_gini:.4f}")
print(f"Recall: {recall_gini:.4f}")
print(f"F1 Score: {f1_gini:.4f}")
accuracy_gini = accuracy_score(y_test, y_pred_gini)
print(f"Accuracy: {accuracy_gini:.4f}")
logloss_gini = log_loss(y_test, y_pred_proba_gini)
print(f"Log Loss: {logloss_gini:.4f}\n")
```

```python
dt_entropy = DecisionTreeClassifier(criterion='entropy', random_state=42)
dt_entropy.fit(X_resampled, y_resampled)
y_pred_entropy = dt_entropy.predict(X_test_scaled)
y_pred_proba_entropy = dt_entropy.predict_proba(X_test_scaled)
print("\n'entropy':")
cm_entropy = confusion_matrix(y_test, y_pred_entropy)
print(f"Confusion Matrix:\n{cm_entropy}")
precision_entropy = precision_score(y_test, y_pred_entropy)
recall_entropy = recall_score(y_test, y_pred_entropy)
f1_entropy = f1_score(y_test, y_pred_entropy)
print(f"Precision: {precision_entropy:.4f}")
print(f"Recall: {recall_entropy:.4f}")
print(f"F1 Score: {f1_entropy:.4f}")
accuracy_entropy = accuracy_score(y_test, y_pred_entropy)
print(f"Accuracy: {accuracy_entropy:.4f}")
logloss_entropy = log_loss(y_test, y_pred_proba_entropy)
print(f"Log Loss: {logloss_entropy:.4f}\n")
```

```python
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [ None, 5, 10, 15, 20, 25,30]
}
dt = DecisionTreeClassifier(random_state=42)
grid_search = GridSearchCV(dt, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
y_pred_proba = best_model.predict_proba(X_test)
print("\nbestDecision Tree :")
cm = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{cm}")
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
logloss = log_loss(y_test, y_pred_proba)
print(f"Log Loss: {logloss:.4f}")
print("\nBest Hyperparameters:", best_params)
```

```python
class_0_weight = 1 / (len(y_resampled[y_resampled == 0]) / len(y_resampled))
class_1_weight = 1 / (len(y_resampled[y_resampled == 1]) / len(y_resampled))
linear_svc = LinearSVC(class_weight={0: class_0_weight, 1: class_1_weight}, random_state=42)
linear_svc.fit(X_resampled, y_resampled)
y_pred_linear = linear_svc.predict(X_test_scaled)
confusion_linear = confusion_matrix(y_test, y_pred_linear)
report_linear = classification_report(y_test, y_pred_linear, target_names=['No Risk', 'Risk'])
print("Kernel: Linear\n")
print("Confusion Matrix:\n", confusion_linear)
print("\nClassification Report:\n", report_linear)
print("\n-------------------------------------------\n")
```

```python
svc_rbf = SVC(kernel='rbf', class_weight={0: 1/len(y_resampled[y_resampled == 0]), 1: 1/len(y_resampled[y_resampled == 1])}, random_state=42)
svc_rbf.fit(X_resampled, y_resampled)
y_pred_rbf = svc_rbf.predict(X_test_scaled)
confusion_rbf = confusion_matrix(y_test, y_pred_rbf)
report_rbf = classification_report(y_test, y_pred_rbf, target_names=['No Risk', 'Risk'])
print("Kernel: RBF\n")
print("Confusion Matrix:\n", confusion_rbf)
print("\nClassification Report:\n", report_rbf)
print("\n-------------------------------------------\n")
```

```python
svc_sigmoid = SVC(kernel='sigmoid', class_weight={0: 1/len(y_resampled[y_resampled == 0]), 1: 1/len(y_resampled[y_resampled == 1])}, random_state=42)
svc_sigmoid.fit(X_resampled, y_resampled)
y_pred_sigmoid = svc_sigmoid.predict(X_test_scaled)
confusion_sigmoid = confusion_matrix(y_test, y_pred_sigmoid)
report_sigmoid = classification_report(y_test, y_pred_sigmoid, target_names=['No Risk', 'Risk'])
print("Kernel: Sigmoid\n")
print("Confusion Matrix:\n", confusion_sigmoid)
print("\nClassification Report:\n", report_sigmoid)
print("\n-------------------------------------------\n")
```

```python
svc_poly = SVC(kernel='poly', class_weight={0: 1/len(y_resampled[y_resampled == 0]), 1: 1/len(y_resampled[y_resampled == 1])}, random_state=42)
svc_poly.fit(X_resampled, y_resampled)
y_pred_poly = svc_poly.predict(X_test_scaled)
confusion_poly = confusion_matrix(y_test, y_pred_poly)
report_poly = classification_report(y_test, y_pred_poly, target_names=['No Risk', 'Risk'])
print("Kernel: Polynomial\n")
print("Confusion Matrix:\n", confusion_poly)
print("\nClassification Report:\n", report_poly)
print("\n-------------------------------------------\n")
```

```python
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
linear_svc = LinearSVC(class_weight={0: 1/len(y_resampled[y_resampled == 0]), 1: 1/len(y_resampled[y_resampled == 1])}, random_state=42)
grid_search = GridSearchCV(linear_svc, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_resampled, y_resampled)
best_C = grid_search.best_params_['C']
```

```python
final_linear_svc = LinearSVC(C=best_C, class_weight={0: 1/len(y_resampled[y_resampled == 0]), 1: 1/len(y_resampled[y_resampled == 1])}, random_state=42)
final_linear_svc.fit(X_resampled, y_resampled)
y_pred_linear_svc = final_linear_svc.predict(X_test_scaled)
report_linear_svc = classification_report(y_test, y_pred_linear_svc, target_names=['No Risk', 'Risk'])
print(f"Best C: {best_C}\n")
print("Classification Report:\n", report_linear_svc)
```
